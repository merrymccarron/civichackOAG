```python
def dump_query_to_csv(q,filename):
    csvfile = open(filename, 'wb') 
    csvwriter = csv.writer(csvfile)

    c = conn.cursor()

    df = pd.read_sql('%s AND rownum < 1' % q ,  con=conn)
    cols = df.columns.values
    print cols
    csvwriter.writerow(cols)

    c.execute(q) 

    for row in c:
        #print row
        csvwriter.writerow(row)

    csvfile.close()
```
